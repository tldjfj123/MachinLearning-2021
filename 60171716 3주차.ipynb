```python
import numpy as np
import  matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data from csv file
data = np.loadtxt('weightHeight.csv', delimiter = ",")

man = []
woman = []

# Classify given data by using for loop
for d in data :
    if d[:1] == 0 : 
        man.append(d[1:])
    else :
        woman.append(d[1:])

# Convert array to numpy array
man = np.asarray(man)
woman = np.asarray(woman)

# Separate data
man_weight = man[:, :1]
man_height = man[:, 1:]
woman_weight = woman[:, :1]
woman_height = woman[:, 1:]

# Convert man's data (2D array to 1D array) for using polyfit method
man_weight = man_weight.ravel()
man_height = man_height.ravel()

# Describe relation between man's weight and height by using polyfit
w = man_weight
h = man_height
k, d = np.polyfit(w, h, 1) 
h_pred = k * w + d
plt.plot(w, h, '*', color='violet')
plt.plot(w, h_pred)
```

```python
# Describe relation between woman's weight and height by using sklearn module
reg = LinearRegression()

ww = woman_weight
wh = woman_height

reg.fit(ww, wh)
whhat = reg.predict(ww) 
plt.plot(ww, wh, 'd')
plt.plot(ww, whhat)
```
