```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def make_img() :
    img = mpimg.imread("C:/Users/SeongJun/Desktop/띵지/기계학습응용/edge_detection_ex.jpg")
    # data format should be changed to batch_type + [height, width, channels]
    img = img.reshape(1, 720, 1280, 3)
    img = tf.constant(img, dtype = tf.float64)
    
    return img

def make_filter() :
    weight = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    weight = np.stack((weight, weight, weight))
    weight = weight.reshape(1, 3, 3, 3)
    weight_init = tf.constant_initializer(weight)
    
    return weight_init
    

def cnn_valid(image, weight) :
    conv2d = tf.keras.layers.Conv2D(filters = 1, kernel_size = (3, 3), padding = "SAME",kernel_initializer=weight)(image)
    plt.imshow(conv2d.numpy().reshape(720, 1280), cmap = 'gray')
    plt.show()   

# def cnn_valid(image) :
#     conv2d = tf.keras.layers.Conv2D(filters = 1, kernel_size = (3, 3), padding = "SAME")(image)
#     plt.imshow(conv2d.numpy().reshape(720, 1280), cmap = 'gray')
#     plt.show()   
    

def main() :
    img = make_img()
    filter = make_filter()
    
    cnn_valid(img, filter)
    # cnn_valid(img)

main()
```
