```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load data from csv file
data = np.loadtxt('weightHeight.csv', delimiter = ",")

man = []

# Classify given data by using for loop
for d in data :
    if d[:1] == 0 : 
        man.append(d[1:])

# Convert array to numpy array
man = np.asarray(man)

# Separate data
man_weight = man[:, :1]
man_height = man[:, 1:]

# Divide test data and train data into 8:2
train_x, test_x, train_y, test_y = train_test_split(man_weight, man_height, test_size=0.2)

# Make LinearRegression Model
reg = LinearRegression()

# Train model by using train set
reg.fit(train_x, train_y)

# Get MSE value by using mean_squared_error function from sklearn
print("Mean Squared Error : {}".format(mean_squared_error(reg.predict(test_x), test_y)))

# Present test value and predicted value
plt.scatter(test_x, test_y, color = 'violet')
plt.plot(test_x, reg.predict(test_x), color = 'blue')
plt.show



```
